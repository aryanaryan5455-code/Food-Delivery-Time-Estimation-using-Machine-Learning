##  Feature Engineering: Develop New Features that could improve model accuracy .

##### Delivery Distance- Most Important Feature

```python
def haversine(lat1, lon1, lat2, lon2):
    R = 6371  # Earth radius in km

    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2*np.arcsin(np.sqrt(a))

    return R * c


train['distance_km'] = train.apply(
    lambda row: haversine(
        row['Restaurant_latitude'],
        row['Restaurant_longitude'],
        row['Delivery_location_latitude'],
        row['Delivery_location_longitude']
    ),
    axis=1
)
```

##### Order Hour Feature

```python
train['Time_Orderd'] = pd.to_datetime(train['Time_Orderd'], errors='coerce')

train['order_hour'] = train['Time_Orderd'].dt.hour
```

##### Order Day of Week

```python
train['Order_Date'] = pd.to_datetime(train['Order_Date'])

train['order_day'] = train['Order_Date'].dt.dayofweek
```

##### Order Preparation Time

```python
train['Time_Order_picked'] = pd.to_datetime(train['Time_Order_picked'], errors='coerce')

train['prep_time'] = (
    train['Time_Order_picked'] - train['Time_Orderd']
).dt.total_seconds() / 60
```

##### Driver Experience Score

```python
train['experience_score'] = (train['Delivery_person_Age'] * train['Delivery_person_Ratings'])
```

##### Delivery Load Feature

```python
train['delivery_load'] = (train['multiple_deliveries'] * train['distance_km'])
```

##### Traffic Level Encoding

```python
traffic_map = {
    'Low':1,
    'Medium':2,
    'High':3,
    'Jam':4
}

train['traffic_level'] = train['Road_traffic_density'].map(traffic_map)
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(seg_scaled)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11), wcss, marker='o')
plt.title("Elbow Method for Optimal Clusters")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")

plt.show()
```

Step 4 Applying K-Means Clustering

```python
kmeans = KMeans(n_clusters=3, random_state=42)

train['customer_segment'] = kmeans.fit_predict(seg_scaled)
```

Step 5 Analyzing the segments

```python
train.groupby('customer_segment')[[
    'distance_km',
    'multiple_deliveries',
    'Delivery_person_Ratings',
    'Time_taken(min)'
]].mean()
```

### Model Comparison

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Store predictions from all models
model_predictions = {
    "Linear Regression": pred_lr,
    "Random Forest": pred_rf,
    "Gradient Boosting": pred_gb,
    "XGBoost": pred_xgb,
    "Stacking Model": pred_stack
}

# Calculate metrics
results = []

for name, preds in model_predictions.items():
    
    r2 = r2_score(y_val, preds)
    mae = mean_absolute_error(y_val, preds)
    rmse = np.sqrt(mean_squared_error(y_val, preds))
    
    results.append([name, r2, mae, rmse])

# Create results dataframe
model_comparison = pd.DataFrame(
    results,
    columns=["Model", "R2 Score", "MAE", "RMSE"]
)

# Sort by best R2
model_comparison = model_comparison.sort_values(by="R2 Score", ascending=False)

print(model_comparison)
```
